### 1. Write a format string (using `str.format()` and `f'string'`) that will take the following four element tuple:

`( 2, 123.4567, 10000, 12345.67)`

and produce:
    
`'file_002 :   123.46, 1.00e+04, 1.23e+04'`

### 2. Write a python function that output the following formatted string based on the input iterable:

`'The n numbers are: i1, i2, i3, ...'`
    
where `n` is the number in the iterable and `i1, i2 ...` are the items inside the iterable. 

for example:

`func([1, 2, 3])` will return `'The 3 numbers are: 1, 2, 3'`

### 6. Write a function that will format each item in an input iterable object into a customized width 

syntax: `func(iterable, width)`

for example:
    
`func([1, 2, 3, 4, 5], 3)`  return `'  1  2  3  4  5'`

`func([1, 2, 3, 4, 5], 2)`  return `' 1 2 3 4 5'`

````python
def myFormatter(myList, width): 
    myString = ""
    seperator = ""
    for i in range(width):
        seperator += " "
    for ii in myList:
        myString += seperator + str(ii)
    return myString

myList = [1, 2, 3, 4, 5, 6, 7]
width = 5
print(myFormatter(myList,width))
````

````python
def myFormatter(myList, width): 
    myItemString = ""
    for i in range(len(myList)):
        myItemString += f'{myList[i]:{width}}'
    return myItemString

myList = [1, 2, 3, 4, 5, 6, 7]
width = 5
print(myFormatter(myList,width))
````

### 7. The USGS Water Data service provides retrival of streamflow data at the USGS stream gages. The download url follows the pattern below. Write a Python program 1) generate the url for a given list of gage numbers; 2) read this url and download the data to csv file (do not use numpy or pandas); 3) based on the downloaded data calculate the monthly stasticstics including maximum, minimum and average of the streamflow data and save it to another csv file. The data need to be formatted to 2 decimal digits. 4) calculate the average annual runoff of each gage (expressed in acre-feet) 


url: https://waterdata.usgs.gov/nwis/dv?&cb_00060=on&format=rdb&site_no=06803495&referred_module=sw&period=&begin_date=2019-02-11&end_date=2020-02-11

where `&site_no=06803495` define which gages are included in this data retrieval; for multiple gages, simply repeat it, e.g. `&site_no=06803495&site_no=06803486`;

the datas are specified by `&period=&begin_date=2018-02-11&end_date=2020-02-11`

You could test your program using the gages `['06803495', '06803486']` for the period between `2000-10-1` and `2019-9-30`


````python
import datetime

# create URL for selected stream gages and data range
myStreamGageNums = ['06803495', '06803486']
dateRange = ['2000-10-10', '2019-9-30']

gageString = ''
for gage in myStreamGageNums:
    gageString += "&site_no=" + gage

myURL = ('https://waterdata.usgs.gov/nwis/dv?&cb_00060=on&format=rdb{}&referred_module=sw&period=&begin_date={}&end_date={}').format(gageString,dateRange[0],dateRange[1])
print(myURL, '\n')

# retrieve data from web page, NOTE: the data is in binary format
from urllib.request import urlopen
myURLdata = urlopen(myURL).readlines()

for i in range(40):
    print(myURLdata[i])

print('\n')

myFile = open("myStreamflowFile.csv",'w')      # open new file with write access
myFileHeaders = ['agency','site_no','dateTime','discharge_cfs','qualification']
myFile.write( ','.join(myFileHeaders)+'\n') # first line

# decode binary data and save to text file
i=0
for line in myURLdata:
    strLine = line.decode('utf8').split()   # decode each line of binary data
    # exclude comments (i.e. lines that start with '#')
    if strLine[0] != "#" and strLine[0] == 'USGS':
        iAgency = myFileHeaders.index('agency')
        iSiteNo = myFileHeaders.index('site_no')
        iDateTime = myFileHeaders.index('dateTime')
        iDischarge = myFileHeaders.index('discharge_cfs')
        iQualification = myFileHeaders.index('qualification')
        myFile.write(f'{strLine[iAgency]},{strLine[iSiteNo]},{strLine[iDateTime]},{strLine[iDischarge]},{strLine[iQualification]}\n')
        #print(datetime.datetime.strptime(strLine[iDateTime],'%Y-%m-%d'))
        #print(type(strLine[iDateTime]))
        #myDateTime = (f'{datetime,datetime.strptime(strLine[iDateTime],'%Y-%m-%d')}')
        #print(myDateTime)
        #myFile.write(f'{x[iAgency]},{x[iSiteNo]}\n')
        #myFile.write(f'{x[iAgency]},{x[iSiteNo]}\n')
        if i<=10:
            i+=1
            print(strLine)
myFile.close()
````

````python
def annualRunoff(myStreamflowData):
    currentSite = ''
    currentYear = ''
    begin = True
    with open(myStreamflowData) as f:
        f.readline()
        siteIdx = 1
        dateIdx = 2
        valueIdx = 3
        for k, dataLine in enumerate(f):
            dataItems = dataLine.split(',')
            
            if dataItems[dateIdx][:4] != currentYear or dataItems[siteIdx] != currentSite:
                currentYear = dataItems[dateIdx][:4]
                currentSite = dataItems[siteIdx]
                key = f'{currentSite},{currentYear}'
                if begin:
                    siteYear = {key:[]}
                    begin = False
                else:
                    siteYear[key] = []
            
            if dataItems[valueIdx].replace('.','0').isnumeric():
                
                #convert from 'cubic feet per second' to 'acre feet per day'
                acreFeet = float(dataItems[valueIdx])*24*60*60/43560
                siteYear[key].append(float(acreFeet))
                

    with open('annualRunoff.csv', 'w') as fw:
        fw.write('siteNumber,year,annual runoff (ac/ft)\n')
        for key in siteYear:
            totalRunoff = sum(siteYear[key])
            fw.write(f'{key}-01-01,{totalRunoff:.2f}\n')
        

annualRunoff('myStreamflowFile.csv')

! head -10 'annualRunoff.csv'
! tail -10 'annualRunoff.csv'
````
